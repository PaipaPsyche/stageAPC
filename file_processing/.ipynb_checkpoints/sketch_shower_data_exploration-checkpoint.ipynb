```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import pandas as pd
from corsikaio import CorsikaParticleFile

from matplotlib import cm
from numpy import log10,sqrt
import numpy as np
from random import random
from matplotlib.colors import LogNorm
from astropy.io import fits
from random import random
from astropy.table import Table

dir_in=''
dir_out='out/'


dictID = { # name , charge, mass(GeV)
    1:["photon",0,0],
    2:["e+",1,0.000511],
    3:["e-",-1,0.000511],
    5:["mu+",1,0.105],
    6:["mu-",-1,0.105],
    13:["neutron",0,0.939],
    14:["proton",1,0.938]
}


def assign_phi(x,y):
    phi = 0
    if(x>0):
        if(y>0):
            phi = np.arctan(y/x)
        else:
            phi = 2*np.pi + np.arctan(y/x)
    elif(x==0):
        phi = (np.pi/2)*np.sign(y) 
    elif(x<0):
        phi = np.pi + np.arctan(y/x)
    return phi%(2*np.pi)

# showers / size (KB) / time (s)
performance_data = [[1,68,8],
                   [10,2286,19],
                   [100,24016,131],
                   [1000,310183,1535]]

performance_data = pd.DataFrame(data=performance_data,columns=["showers","size_KB","time_s"])
```

### Particle events 

For each shower, thousands of simulated particles hit the ground and their properties are stored in an object collection where eahcobject corresponds to a particle. wtih the momentum data, the impact angle $\theta$ , the azimuthal angle $\phi$ and energy can be reconstructed. To save memory, in the furure only position, energy and type data will be stored.   

```python
f=CorsikaParticleFile('DAT000002') #File input (format ?)

events = []

tot_ev = 0

shower_count = 0
for e in f:
    shower_count+=1
    typp=e.particles['particle_description']
    typ=np.trunc(np.array(typp)/1000.)
    
    # momentum in GeV
    px=e.particles['px']
    py=e.particles['py']
    pz=e.particles['pz']
    
    x=e.particles['x']/1.e2
    y=e.particles['y']/1.e2
    
    t=e.particles['t']
    
    lev=np.zeros(len(typ))
    lev=lev.astype(int)  
    
    
    # E2 = p2 + m2
    #mass in GeV
    
    for i in range(len(typ)): 
        code = typ[i]
        
        if(code in dictID.keys()):
            particle = dictID[code]
            E = sqrt(px[i]**2+py[i]**2+pz[i]**2+particle[2]**2)
            event = {
                "event":shower_count,
                "particle":particle[0],
                "charge":particle[1],
                "mass":particle[2],
                "energy":E,
                "log_energy":np.log10(E),
                "px":px[i],
                "py":py[i],
                "pz":pz[i],
                "momentum":sqrt(px[i]**2+py[i]**2+pz[i]**2),
                "theta":np.arccos(pz[i]/sqrt(px[i]**2+py[i]**2+pz[i]**2)),
                "phi":assign_phi(px[i],py[i]),
                "x":x[i],
                "y":y[i],
                "time":t[i]
            }
            events.append(event)
        tot_ev+=1

events = pd.DataFrame(events)

```
